# Problem2: Define a function to add new row data to a csv file .

```python
def addRowCSV(filename, rowdata):
    df=readCSV(filename)
    df.loc[len(df)+1]=rowdata
    df.to_csv(filename,index=False)
    return df
rowdata=['010100','priya',123,4323,6756,1234,1234,345,2345,123,567]
addRowCSV('Datafiles/income.csv',rowdata)
```

# Problem 3: Define a function to update/modify a row in a csv file data.

```python

def updateRowCSV(filename,rowindex,rowdata):
    df=readCSV(filename)
    df.loc[rowindex]=rowdata
    df.to_csv(filename,index=False)
    
    return df

rowdata=[121,'saranya',24234,454,432,213,2434,5465,243,354,323]

updateRowCSV('Datafiles/income.csv',5,rowdata)
    


    
    
```

# Problem 5: Define a function to delete a row from csv file

```python
def deleteRowCSV(filename,rowindex):
    df=readCSV(filename)
    
    df=df.drop(rowindex)
    df.to_csv(filename,index=False)
    return df

deleteRowCSV('Datafiles/income.csv',4)    
```
