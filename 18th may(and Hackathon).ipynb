# Problem 5: Generate a Histogram for data.

```python
#!/usr/bin/env python
 
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
 
 
# example data
mu = 100 # mean of distribution
sigma = 15 # standard deviation of distribution
x = mu + sigma * np.random.randn(10000)
 
num_bins = 20
# the histogram of the data
n, bins, patches = plt.hist(x, num_bins, normed=1, facecolor='blue', alpha=0.5)
 
# add a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')
 
# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
plt.show()
```

```python
# method 2 for Pie Chart to categorise all app categories based on count

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('Datafiles/AppleStore.csv')
val=[]
labels=[]
sizes=[]
count={}
for i in df.values:
    a=i[12]
    val.append(a)
for element in val:
    count[element] = count.get(element, 0) + 1
#print(count)

for label in count:
    labels.append(label)
    
    
for size in count.values():
    sizes.append(size)
    
    
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','lightcoral']

plt.pie(sizes,  labels=labels,  colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()

```
